Compare most popular tags in year 2016 with tags in 2009. Tag popularity is the number of questions (post_type_id=1) with this tag. Output top 10 tags in format:

tag rank in 2016 rank in 2009 tag popularity in 2016 tag popularity in 2009

Using WITH helps to make query code clean and readable.

```python
%%writefile query.hql

ADD JAR /opt/cloudera/parcels/CDH/lib/hive/lib/hive-contrib.jar;
ADD JAR /opt/cloudera/parcels/CDH/lib/hive/lib/hive-serde.jar;

USE stackoverflow_;

SELECT tag, year, post_type_id, count(*) as popularity_tag
FROM posts
LATERAL VIEW explode (tags)lateral_table AS tag
WHERE year = '2009' OR year='2016'
GROUP BY tag, year, post_type_id;
```
